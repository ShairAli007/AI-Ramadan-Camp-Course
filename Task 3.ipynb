```python
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. Power")

choice = input("Enter choice(1/2/3/4/5): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
if choice == '1':
    print(f"{num1} + {num2} = {num1 + num2}")
elif choice == '2':
    print(f"{num1} - {num2} = {num1 - num2}")
elif choice == '3':
    print(f"{num1} * {num2} = {num1 * num2}")
elif choice == '4':
    print(f"{num1} / {num2} = {num1 / num2}")
elif choice == '5':
    print(f"{num1} ** {num2} = {num1 ** num2}")
else:
    print("Invalid Input")
```

```python
my_list = ['apple', 10, 'banana', 25.5, 'cherry']
numeric = False
for item in my_list:
    if isinstance(item, (int, float)):
        numeric = True
        break        
if numeric:
    print("The list contains a numeric value.")
else:
    print("The list does not contain a numeric value.")
```

```python
my_dict = {'brand': 'Ford', 'model': 'Mustang', 'year': 1964}
my_dict['color'] = 'red'
print(my_dict)
```

```python
data_dict = {'item1': 10, 'item2': 'hello', 'item3': 20.5, 'item4': 'world'}
total_sum = 0
for value in data_dict.values():
    if isinstance(value, (int, float)):
        total_sum += value
print(f"The sum of numeric items is: {total_sum}")
```

```python
sample_list = [1, 2, 3, 2, 1, 4, 5, 3]
seen = set()
duplicates = set()
for item in sample_list:
    if item in seen:
        duplicates.add(item)
    else:
        seen.add(item)
print(f"Duplicate values: {list(duplicates)}")
```
