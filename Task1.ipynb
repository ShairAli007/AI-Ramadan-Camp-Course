```python
num1=float(input ("Enter num1: "))
num2=float(input ("Enter num2: "))
opr= input("Enter operator: ")
if opr== "+":
   print(f"Addition of {num1} + {num2} = {num1+num2}")
elif opr=="-":
   print(f"Subtition of {num1} {num2} = {num1-num2}")
elif opr=="*":
   print(f"Multiplication of {num1} * {num2} = {num1*num2}")
elif opr=="/":
   print(f"Division of {num1} / {num2} = {num1/num2}")
elif opr=="%":
   print(f"Model of {num1} % {num2} = {num1%num2}")
else: print ("invalid operator")
```

```python
Eng = int(input("Enter English Marks:"))
Isl = int(input("Enter Isl Marks:"))
Math = int(input("Enter Maths Marks:"))
Urdu = int(input("Enter Urdu Marks:"))
Comp = int(input("Enter Comp Marks:"))
obt = Eng + Isl + Math + Urdu + Comp
print("Your Obtain Marks is ", obt)
per = obt / 500 * 100
if per <= 100 and per >= 90:
   grade = "A+"
elif per <= 89 and per >= 80:
   grade = "A"
elif per <= 79 and per >= 70:
   grade = "B"
elif per <= 69 and per >= 60:
   grade = "C"
elif per <= 59 and per >= 50:
   grade = "D"
else:
   grade = "F"
print("Percentage:", per)
print("Grade:", grade)
```
